```python
import numpy as np
import matplotlib.pyplot as plt
```

# Method Logic

```python
def plot_1DImage(x_values,y_values, fname = None):

    plt.figure(figsize=(len(x_values)-1, 1.6) )
    plt.fill_between(x_values, 0, y_values, step = "pre")
    
    plt.xlabel("Pixel Location", fontsize =14)
    plt.ylabel("Intensity", fontsize =14)
    
    plt.xticks(range(len(x_values)),range(len(x_values)))
    plt.yticks([min(y_values), max(y_values)], [min(y_values), max(y_values)])
    plt.xlim(0, len(x_values)-1)
    plt.ylim( min(y_values), max(y_values) )
    plt.grid()
    plt.tight_layout()
    
    if fname != None:
        plt.savefig(fout+fname)
        print("Saved Figure as: '{}'".format(fout+fname))
    plt.show()
    
```

```python
def Step_1D(ImageLength, EdgeLocation, startIntensity = 0, endIntensity = 1):
    x_values = [i for i in range(0, ImageLength+1)]
    y_values = [startIntensity if x <= EdgeLocation else endIntensity for x in x_values]
    
    return (x_values, y_values)

def diffmat(matrix0, matrix1):
    diff_matrix = [vals[0] - vals[1] for vals in zip(matrix0, matrix1)]
    return diff_matrix

def plot_diffmat(x_values, y_values, fname = None):
    
    plt.figure( figsize=( len(x_values)-1, 2))
    plt.step(x_values, y_values)
    
    plt.xlabel("Pixel Location", fontsize =14)
    plt.ylabel("Intensity\nDifference", fontsize =14)
    
    plt.yticks([-1, 0, 1], [-1, 0, 1])
    plt.ylim( -1, 1 )
    plt.tight_layout()
    
    
    if fname != None:
        plt.savefig(fout+fname)
        print("Saved Figure as: '{}'".format(fout+fname))
    plt.show()
```

```python
fout = "Results/1D_Examples/"

ImageLength = 6
EdgeLocation = 3
startIntensity = 1
endIntensity = 0

print("\n\n\nOriginal Image")
xval_original, yval_original = Step_1D(ImageLength, EdgeLocation, startIntensity, endIntensity)
plot_1DImage(xval_original, yval_original,"OriginalImage.png")



print("1px Translated Image:")
xval_t1, yval_t1 = Step_1D(ImageLength, EdgeLocation+1, startIntensity, endIntensity)
plot_1DImage(xval_t1, yval_t1, "1pxTranslatedImage.png")

print("Difference Image:")
yval_Ot1 = diffmat(yval_t1, yval_original)
plot_diffmat(xval_original, yval_Ot1, "1pxT_DiffImage.png")
sum_diff = sum(yval_Ot1) #sum of difference
print("Sum of Difference: {}".format(sum_diff))
print("\n\n")



print("2px Translated Image:")
xval_t2, yval_t2 = Step_1D(ImageLength, EdgeLocation+2, startIntensity, endIntensity)
plot_1DImage(xval_t1, yval_t2, "2pxTranslatedImage.png")

print("Difference Image:")
yval_Ot2 = diffmat(yval_t2, yval_original)
plot_diffmat(xval_original, yval_Ot2, "2pxT_DiffImage.png")
sum_diff = sum(yval_Ot2) #sum of difference
print("Sum of Difference: {}".format(sum_diff))
print("\n\n")



print("-1px Translated Image:")
xval_t1, yval = Step_1D(ImageLength, EdgeLocation-1, startIntensity, endIntensity)
plot_1DImage(xval_t1, yval, "-1pxTranslatedImage.png")

print("Difference Image:")
yval = diffmat(yval, yval_original)
plot_diffmat(xval_original, yval, "-1pxT_DiffImage.png")
sum_diff = sum(yval) #sum of difference
print("Sum of Difference: {}".format(sum_diff))
print("\n\n")
```

```python
def MaxTranslationTest(ImageLength, EdgeLocation, startIntensity, endIntensity, blocksize, dx):
    print("{}px T MaxTranslation Test".format(dx))
    xval_dx, yval_dx = Step_1D(ImageLength+dx, EdgeLocation+dx-blocksize, startIntensity, endIntensity)

    discard, yval2 = Step_1D(ImageLength-1-dx, EdgeLocation+dx, startIntensity, endIntensity)
    yval_dx += yval2
    xval_dx = range(len(yval_dx))

    plot_1DImage(xval_dx, yval_dx,"{}pxT_MaxT.png".format(dx))

    print("Difference Image:")
    yval_Ot1 = diffmat(yval_dx, yval_original)
    plot_diffmat(xval_dx, yval_Ot1, "{}pxT_MaxT_DiffImage.png".format(dx))
    sum_diff = sum(map(abs,yval_Ot1)) #sum of difference
    print("Sum of Difference: {}".format(sum_diff))
    print("\n\n")

#----------------


ImageLength = 3
EdgeLocation = 3
startIntensity = 0
endIntensity = 1

blocksize = 2

dx = 0

print("\n\n\nOriginal Image")
xval_original, yval_original = Step_1D(ImageLength, EdgeLocation+dx-blocksize, startIntensity, endIntensity)

discard, yval2 = Step_1D(ImageLength-1, EdgeLocation+dx, startIntensity, endIntensity)

yval_original += yval2
xval_original = range(len(yval_original))
plot_1DImage(xval_original, yval_original,"OriginalImage_MaxT.png")

#----------------
dx+=1
MaxTranslationTest(ImageLength, EdgeLocation, startIntensity, endIntensity, blocksize, dx)
dx+=1
MaxTranslationTest(ImageLength, EdgeLocation, startIntensity, endIntensity, blocksize, dx)
dx+=1
MaxTranslationTest(ImageLength, EdgeLocation, startIntensity, endIntensity, blocksize, dx)
# dx+=1
# MaxTranslationTest(ImageLength, EdgeLocation, startIntensity, endIntensity, blocksize, dx)
```

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

mu = 0
sig=0.2

x_values = np.arange(-1,1.01,0.02)
x_values_2 = x_values-0.2

fx = 1*gaussian(x_values, mu, sig)
gx = 1*gaussian(x_values_2, mu, sig)

fig = plt.figure(figsize=(4,3), dpi=300)
ax = fig.add_subplot(111)
    
ax.plot(x_values, fx, 'o', label="Gaussian, f(x)", markersize=1)
ax.plot(x_values, gx, 'o', label="Gaussian shifted +0.2 samples, f(x+0.2)", markersize=1)

ax.set_ylabel('Magnitude')
ax.set_xlabel('X Sample')

ax.set_xlim(-1,1)

plt.tight_layout()

fig.legend(loc="lower center", ncol=1)
plt.subplots_adjust(bottom=0.35) #give some extra room for the legend
plt.savefig(fout+"CorrelationMethodGaussian_1.png")
plt.show()
```
